### Задание 2
Получите датафрейм authors_price, соединив датафреймы authors и books по полю author_id.


```python
#ваш код
table1 =pd.DataFrame(  
                columns=['author_id', 'author_name'],
                 )
table1['author_id']=[1, 2, 3]
table1['author_name'] =  ['Тургенев', 'Чехов', 'Островский']
#print('table 1: ',table1)

table2 = pd.DataFrame(columns=['author_id','book_title', 'price'])
table2['author_id'] =  [1, 1, 1, 2, 2, 3, 3]
table2['book_title'] = ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники']
table2['price'] =  [450, 300, 350, 500, 450, 370, 290]
#print('table 2:',table2)

table_concat=pd.merge(table1, table2, on='author_id')
table_concat
```

### Задание 3
Создайте датафрейм top5, в котором содержатся строки из authors_price с пятью самыми дорогими книгами.

```python
#ваш код
a = table_concat.sort_values('price', ascending=False)
a.head()
```
