## Задание 1
Напишите программу, которая удаляет дубликаты из списка (в этой задаче нельзя пользоваться множествами).

```python
test_list = [2, 3, 6, 8, 19, 21, 22, 3, 3, 5, 20, 21]

# Решение ниже
resultantList = []
 
for element in test_list:
    if element not in resultantList:
        resultantList.append(element)

print(resultantList)
```

## Задание 2
Напишите программу, которая в каждом подлисте находит максмум и суммирует их.

```python
test_list = [[1,2,3], [4,5,6], [10,11,12], [7,8,9]]
sum = 0
for x in test_list:
  print("max in this list",max(x))
  for y in x:
    sum  = y + sum
  print("sum in this list",sum)
```

## Задание 3
Есть список x. Нужно посчитать количество нечетных элементов с использованием метода .pop() .pop() дает возможность одновременно удалить элемент (список занимает меньше места в памяти) и сохранить его в переменной.

```python
x = [1,2,3,4,5,6,7]
print(len([elem for elem in x if elem % 2 != 0]))
```
